## Section 1.2: BAG OF WORDS

A bag-of-words model, or BoW for short, is a way of extracting features from text for use in modeling, such as with machine learning algorithms.

The approach is very simple and flexible, and can be used in many ways for extracting features from documents.

A bag-of-words is a representation of text that describes the occurrence of words within a document.
It is called a “bag” of words, because any information about the order or structure of words in the document is discarded. The model is only concerned with whether known words occur in the document, not where in the document.

```python
# Functions to convert document(s) to a list of words, with the option of removing stopwords. Returns document-term matrix.

def createBagOfWords(train, test, remove_stopwords, lemmatize, stemmer):
    if remove_stopwords:
        vectorizer = CountVectorizer(analyzer='word', input='content', stop_words=stopwords.words('english'))
    else:
        vectorizer = CountVectorizer(analyzer='word', input='content')

    clean_train = []
    for paragraph in train:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_train.append(paragraph)

    clean_test = []
    for paragraph in test:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_test.append(paragraph)

    bag_of_words_train = vectorizer.fit_transform(clean_train).toarray()
    bag_of_words_test = vectorizer.transform(clean_test).toarray()
    return bag_of_words_train, bag_of_words_test
```

## Section 1.3: TF-IDF
TF-IDF technique is used to find meaning of sentences consisting of words and cancels out the incapabilities of Bag of Words technique which is good for text classification or for helping a machine read words in numbers.

The number of times a term occurs in a document is called its Term frequency (TF).

 Document frequency is the number of documents in which the word is present.  Inverse DF (IDF) is the inverse of the document frequency which measures the informativeness of term *t*.




```python
def createTFIDF(train, test, remove_stopwords, lemmatize, stemmer):
    if remove_stopwords:
        vectorizer = TfidfVectorizer(analyzer='word', input='content', stop_words=stopwords.words('english'))
    else:
        vectorizer =  TfidfVectorizer(analyzer='word', input='content')

    clean_train = []
    for paragraph in train:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_train.append(paragraph)

    clean_test = []
    for paragraph in test:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_test.append(paragraph)

    tfidf_train = vectorizer.fit_transform(clean_train).toarray()
    tfidf_test = vectorizer.transform(clean_test).toarray()
    return tfidf_train, tfidf_test
```

### Questions to Think About and Answer
1. Why does the TF-IDF approach generally result in a better accuracy than Bag-of-Words
The TF-IDF (Term Frequency-Inverse Document Frequency) approach typically results in better accuracy than the Bag-of-Words (BoW) model because it addresses some of the limitations inherent in BoW. Here's how:

Weighting Words by Importance:

In BoW, every word is treated as equally important, and the model counts how often each word appears in the document. This approach doesn't account for the fact that some words (like "the", "is", "on") appear very frequently in the language but are not meaningful for distinguishing between documents. These common words are not helpful for classification or information retrieval.
TF-IDF, on the other hand, adjusts for the frequency of a word in a document (TF) and the rarity of the word across all documents (IDF). Words that appear frequently in a document but not in many others are given higher weight, which helps to highlight the most important, distinctive words in each document. As a result, TF-IDF improves the model’s ability to capture the unique characteristics of a document and often leads to better accuracy in tasks like text classification, search, and clustering.
Reducing the Impact of Common Words:

Words that appear in many documents (such as "the", "and", "in") are weighted less in the TF-IDF model, so they don't dominate the classification or analysis. This allows the model to focus on more meaningful, discriminative words that are relevant for identifying specific topics or themes.
Improved Feature Representation:

In BoW, a document is represented purely as a vector of word counts, leading to sparse and potentially noisy feature vectors. TF-IDF gives a more nuanced representation, where the importance of words is taken into account, improving the accuracy and interpretability of the features.
2. Can you think of techniques that are better than both BoW and TF-IDF ?
Yes, there are several advanced techniques that can often outperform both BoW and TF-IDF, especially in tasks requiring better semantic understanding and context. These methods generally move beyond simple word counting and consider the relationships between words, sentence structure, or context in the document. Some of these techniques include:

Word Embeddings (e.g., Word2Vec, GloVe, FastText):

Word embeddings like Word2Vec or GloVe represent words in a continuous vector space, capturing semantic relationships between words. Similar words (e.g., "cat" and "dog") are located closer to each other in the vector space. This allows the model to understand synonyms and context better than TF-IDF or BoW, which treat words independently.
Embeddings are pre-trained on large corpora and provide richer, more nuanced features than BoW or TF-IDF, which are simply based on word counts and frequencies.
Contextualized Word Embeddings (e.g., BERT, GPT):

Models like BERT and GPT represent words not just as static vectors (as in Word2Vec), but as dynamic vectors that change depending on the context of the sentence. This is crucial because the meaning of a word can vary based on surrounding words. For instance, "bank" means different things in the context of "river bank" vs. "savings bank." BERT's bidirectional nature allows it to consider the full context of a word, which makes it more powerful for tasks like sentiment analysis, text classification, and information extraction.
Doc2Vec:

Doc2Vec is an extension of Word2Vec that generates fixed-size vector representations for entire documents, rather than individual words. This is useful when the meaning of the document as a whole matters more than the individual words. It is particularly useful in document classification, topic modeling, and retrieval tasks where the semantic meaning of the entire document needs to be captured.
Transformer-based Models:

Transformer models, like BERT, RoBERTa, and XLNet, provide state-of-the-art performance for many NLP tasks. These models learn contextual dependencies between words in a sentence and can capture complex linguistic features, such as syntax and semantics, in ways that BoW and TF-IDF cannot. They also support fine-tuning on specific tasks to further improve performance
3. Read about Stemming and Lemmatization from the resources given below. Think about the pros/cons of each.
Both stemming and lemmatization are techniques used to reduce words to their base or root form, helping to standardize variations of words (e.g., "running" → "run"). However, they differ in how they do this, and each has its own advantages and limitations.

Stemming:
Pros:

Faster: Stemming algorithms (like the Porter Stemmer) are typically faster than lemmatization because they use simple heuristics and rule-based approaches to chop off word endings.
Simple: It’s a straightforward approach that often works well for improving retrieval or classification tasks by reducing different forms of a word to a common root.
Effective in Some Tasks: In certain applications like search or keyword extraction, stemming can still provide useful results, especially when the goal is to match words at a basic level.
Cons:

Over-stemming: Stemming often produces "root" forms that are not valid words (e.g., "running" becomes "run," but "flying" becomes "fli"), which can lead to loss of meaningful distinctions between words.
Loss of Meaning: Stemming doesn't always preserve the semantic meaning of the word. For example, “better” might be stemmed to “bet,” which distorts the actual meaning.
Lemmatization:
Pros:

Preserves Meaning: Lemmatization reduces words to their base form (lemma) based on their correct meaning and grammatical context. For example, "running" becomes "run," and "better" becomes "good." This makes lemmatization more semantically accurate than stemming.
More Accurate: Since lemmatization considers the part of speech of a word (noun, verb, etc.), it results in more natural, meaningful words that are better suited for downstream tasks like text classification or sentiment analysis.
Cons:

Slower: Lemmatization is computationally more expensive than stemming because it requires a dictionary or a more sophisticated linguistic algorithm to look up the proper lemma.
Requires POS Tagging: To lemmatize correctly, you often need part-of-speech tagging, which requires additional preprocessing and can be error-prone.
In Summary:
Stemming is fast and simple but can produce non-standard words that may lose meaning or distort the text.
Lemmatization is more accurate and preserves meaning but is computationally more expensive and may require additional processing steps like POS tagging.
For most modern NLP applications, lemmatization is generally preferred when accuracy is important, though stemming might still be useful in some cases where speed is a priority and perfect semantic understanding is not critical.
